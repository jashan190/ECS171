```python
import numpy as np
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.datasets import fetch_california_housing

def polynomial_regression(degree, X, y, folds, test_size=0.25, random_state=None):
    # Define number of folds for cross-validation
    kf = KFold(folds)

    # Initialize lists to store results for variance, bias2s, total_error, and models
    # TODO
    variance_results = []
    bias2s_results = []
    total_error= []
    models= []


    # Set the polynomial degree of the model
    poly_features = PolynomialFeatures(degree)
    X_poly = poly_features.fit_transform(X)

    # Perform cross-validation
    for train_index, test_index in kf.split(X_poly):
        # Split data into training and testing sets for this fold
        X_train, X_test = X_poly[train_index], X_poly[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # Fit polynomial regression model
        model = LinearRegression()
        model.fit(X_train, y_train)

        # Make predictions on the test set
        y_pred = model.predict(X_test)

        # Calculate variance and bias for this fold
        ## TODO
        variance = np.var(y_pred)

        bias2 = np.mean((y_test - y_pred) ** 2)
        
        # Append results to lists
        # TODO
        variance_results.append(variance)
        bias2s_results.append(bias2)
        models.append(model)
        total_error.append(bias2 + variance)

        # Print results for this fold
        print("Variance: {:.4f}, Bias2: {:.4f}, Total error: {:.4f}".format(variance, bias2, bias2 + variance))

    # print the total_error of the best model
    min_error_index = np.argmin(total_error)
    best_model = models[min_error_index]
    # TODO
    print("The Best model is: (degree {}): And total_error is: {:.4f}".format(degree,total_error[min_error_index]))
    # Testing the final model on the test data
    X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=test_size, random_state=42)
    # Obtain the predictions on the test data
    # TODO
    y_pred_test = best_model.predict(X_test)
    # store mse score of the model applied on the test data
    # TODO
    mse = mean_squared_error(y_test, y_pred_test)

    return mse , best_model






# Example usage: load California Housing Dataset and select the first, third, and forth attributes as input features in X
# TODO
houses = fetch_california_housing()
#Professor says put 1, 3, an4 as input
X = houses.data[:, [1,3,4]]
# Set the target valiable 
# TODO
y = houses.target


degrees = range(1, 6)  # Try polynomial degrees from 1 to 5
# Try degrees from 1 to 5 and in a loop, report mse of the best model trained using k-fold cross validation and print("Degree:", degree, "MSE:", mse)
# TODO

for degree in degrees:
    mse, model = polynomial_regression(degree,X,y, folds = 5)

    




    print("Degree: ",degree, "MSE: ", mse)

```

As we increase the degree of the polynomial, we start to see the usual tradeoff between bias and variance. When the model is too simple—like with degree 1 or 2—it tends to miss important patterns, showing higher bias and underfitting the data. On the flip side, higher-degree models (like degree 4 or 5) do a better job capturing the training data, which lowers bias, but they also become more sensitive to noise, leading to higher variance and a risk of overfitting.

Degree 3 strikes a good balance. It keeps the total error stable and gives a relatively low mean squared error (MSE), making it a solid choice. While degree 5 does produce the lowest test MSE overall, its performance is less consistent across different data splits, which suggests it might not generalize as well.
