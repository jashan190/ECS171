Q1 (10 points) Use the sklearn Breast_cancer dataset and use min-max scalar to transform the input attributes. Next, develop two classifiers using logistic regression, and perceptron learning. Train on the training data (75% of the entire data) and compare the performance of the models by reporting accuracy "accuracy = accuracy_score(y_test, y_pred). Which model performs better? Provide your coding for the developed models and document your code. Failing proper documentation leads to losing points. Necessary library functions are provided.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression, Perceptron
from sklearn.metrics import accuracy_score
```

```python
#load data
dataset = load_breast_cancer()
X = dataset.data
y = dataset.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
scaler = MinMaxScaler()
trained_X = scaler.fit_transform(X_train)
#now develop two classifiers 
Logistic_model = LogisticRegression()
Perceptron_model = Perceptron()
#train the models
Logistic_model.fit(trained_X, y_train)
Perceptron_model.fit(trained_X, y_train)
#test the models
trained_X_test = scaler.transform(X_test)
Logistic_pred = Logistic_model.predict(trained_X_test)
Perceptron_pred = Perceptron_model.predict(trained_X_test)
#calculate accuracy
Logistic_accuracy = accuracy_score(y_test, Logistic_pred)
Perceptron_accuracy = accuracy_score(y_test, Perceptron_pred)
print(f"Logistic Regression Accuracy: {Logistic_accuracy}")
print(f"Perceptron Accuracy: {Perceptron_accuracy}")
# The code above loads the breast cancer dataset, splits it into training and testing sets, scales the features using MinMaxScaler, trains two classifiers (Logistic Regression and Perceptron), and evaluates their accuracy on the test set. The accuracies are printed at the end.
```

Q3 (8 points) Mathematically explain how a perceptron learning model is trained. Discuss in terms of the gradient of the error function used in Perceptron Learning algorithm. 

A perceptron model is trained through binary classification  meaning the output of it will either be a 0 or 1. This is the simplest form of an ANN (artifical neural network). 

\[
\hat{y} = \text{step}(w \cdot x)
\]

The training process uses the Perceptron Learning Rule:

\[
w := w + \alpha(y - \hat{y})x
\]

- \( w \) are the weights
- \( x \) is the input vector
- \( y \) is the true label
- \( \hat{y} \) is the predicted label
- \( \alpha \) is the learning rate

The Perceptron Learning algorithm uses a binary step activation function that outputs either 0 or 1 based on whether the weighted sum of inputs crosses a threshold. In contrast, linear regression produces a continuous output and can be used for binary classification by applying a threshold (e.g., 0.5). However, neither approach supports nonlinearity in data. The perceptron can only classify data that is linearly separable, and its step function is non-differentiable, limiting its learning capabilities. Linear regression assumes a linear relationship between inputs and outputs, which also restricts it to linearly separable problems. As a result, both methods fail to capture complex patterns in nonlinear datasets without additional feature transformations or the use of nonlinear activation functions.


Q5 (14 points) Answer the following questions: 
- (8 points) Discuss the vanishing gradient problem in the context of training deep neural networks and identify activation functions that are particularly susceptible to this phenomenon. 
  In deep neural networks, gradients are computed layer by layer using backpropagation. When activation functions like sigmoid or tanh are used, the gradients can become very small as they are multiplied across layers.

This leads to the **vanishing gradient problem**, where early layers receive negligible updates, causing the network to train very slowly or not at all.

**Activation functions prone to vanishing gradients**:
- **Sigmoid**: derivative maxes out at 0.25
- **Tanh**: derivative maxes out at 1 but quickly approaches 0 for large inputs

- (6 points) Explain why these activation functions lead to vanishing gradients during backpropagation (hint: discuss in terms of the shape of the activation function). 
The sigmoid and tanh functions "saturate" at extreme input values. That is, for large positive or negative inputs, their output approaches a flat line (0 or 1 for sigmoid, -1 or 1 for tanh).

Because the slope (derivative) of the function is near **zero** in these regions, the gradient passed back during training is also near zero:

\[
\sigma'(x) = \sigma(x)(1 - \sigma(x)) \approx 0 \quad \text{for large } |x|
\]

This leads to tiny weight updates in earlier layers, preventing the network from learning effectively — a key characteristic of the vanishing gradient problem.



Q6 (12 points) List all hyperparameters discussed in the class related to Artificial and Deep Neural Networks and explain the role/impact of each hyperparameter. Which technique(s) can be used to perform hyperparameter tuning? Explain how the technique(s) work.  

Hyperparameters in Artificial and Deep Neural Networks determine the structure and training behavior of the model. Below are key hyperparameters discussed in class and their roles:

- **Learning Rate (α):** Controls how much the weights are updated during training. A small value leads to slow convergence, while a large one may overshoot the minimum.
- **Number of Epochs:** Refers to how many times the model will see the entire dataset. Too few can lead to underfitting, too many to overfitting.
- **Batch Size:** The number of samples processed before updating the model. Affects training speed and stability (smaller batches introduce more noise).
- **Number of Hidden Layers:** Increasing layers allows the model to learn more complex patterns but also increases computation and risk of overfitting.
- **Number of Neurons per Layer:** More neurons can model more complex relationships but may cause overfitting if excessive.
- **Activation Function:** Determines the output of each neuron. Common choices include ReLU, sigmoid, tanh. Nonlinear activations enable learning non-linear patterns.
- **Weight Initialization:** Affects how quickly and well the network converges. Poor initialization can lead to vanishing or exploding gradients.
- **Momentum:** Helps accelerate gradient descent by dampening oscillations in directions with high curvature.
- **Regularization Parameter (e.g., L2 penalty):** Penalizes large weights to prevent overfitting.

**Hyperparameter Tuning Techniques:**
- **Grid Search:** Try all combinations of predefined hyperparameter values.
- **Random Search:** Randomly sample hyperparameters, often more efficient than grid search.
- **Bayesian Optimization:** Uses probabilistic models to choose the next best set of hyperparameters.
- **Automated Tools:** Libraries like Optuna, Hyperopt, or Keras Tuner automate this process using advanced heuristics.


Q7 (20 points) Given a dataset with input attributes x1 and x2, and output variable y, you are training a 3 layer neural network. Assume that activation function used in each layer is sigmoid. Mathematically describe one feed-forward pass followed by one backward-pass in terms of updating the weights of each layer in this neural network. 

Let the network have an input layer with attributes \( x_1, x_2 \), one hidden layer (layer 1), and one output neuron (layer 2). Let \( \sigma(x) = \frac{1}{1 + e^{-x}} \) be the sigmoid activation function.

#### **Feed-forward pass:**

1. **Hidden layer net input:**

\[
z^{(1)} = W^{(1)} \cdot X + b^{(1)}
\]

\[
a^{(1)} = \sigma(z^{(1)})
\]

2. **Output layer net input:**

\[
z^{(2)} = W^{(2)} \cdot a^{(1)} + b^{(2)}
\]

\[
\hat{y} = a^{(2)} = \sigma(z^{(2)})
\]

---

#### **Backward pass:**

1. **Output layer error (using Mean Squared Error):**

\[
\delta^{(2)} = (\hat{y} - y) \cdot \sigma'(z^{(2)}) = (\hat{y} - y) \cdot \hat{y}(1 - \hat{y})
\]

2. **Hidden layer error:**

\[
\delta^{(1)} = \left(W^{(2)}\right)^T \cdot \delta^{(2)} \cdot \sigma'(z^{(1)})
\]

3. **Weight and bias updates:**

\[
W^{(2)} := W^{(2)} - \alpha \cdot \delta^{(2)} \cdot a^{(1)}^T
\]

\[
W^{(1)} := W^{(1)} - \alpha \cdot \delta^{(1)} \cdot X^T
\]

\[
b^{(2)} := b^{(2)} - \alpha \cdot \delta^{(2)}
\quad , \quad
b^{(1)} := b^{(1)} - \alpha \cdot \delta^{(1)}
\]

This completes one full forward and backward pass of training in a 3-layer neural network using sigmoid activation.


Q8 (20 points) In this exercise, use the output information generated by this code to perform a comparative study of the performance (i.e., loss , Accuracy) of the neural networks models based on the hyperparameters used. Generate a table to report your analysis.
Note: If you want to run this code and have trouble with imported libraries, try 'pip install keras==2.12.0'

```python
import numpy as np
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

# Load dataset
digits = load_digits()
X, y = digits.data, digits.target

# Preprocess data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Define create_model function for KerasClassifier
def create_model(num_layers=1, num_neurons=64, activation='relu', dropout_rate=0.0, momentum=0.9):
    model = Sequential()
    model.add(Dense(num_neurons, input_dim=X_scaled.shape[1], activation=activation))
    model.add(Dropout(dropout_rate))
    for _ in range(num_layers - 1):
        model.add(Dense(num_neurons, activation=activation))
        model.add(Dropout(dropout_rate))
    model.add(Dense(10, activation='softmax'))
    optimizer = SGD(learning_rate=0.01, momentum=momentum)
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

# Define parameters grid for grid search
param_grid = {
    'num_layers': [3],
    'num_neurons': [32, 64],
    'activation': ['relu', 'tanh'],
    'dropout_rate': [0.2, 0.5],
    'momentum': [0.5, 0.9]
}

# Create KerasClassifier wrapper for scikit-learn
model = KerasClassifier(build_fn=create_model, epochs=10, batch_size=32)

# Perform grid search with cross-validation
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=kfold, scoring='accuracy')
grid_result = grid_search.fit(X_scaled, y)

# Print results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, std, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, std, param))
```
